```python
import random

data1 = data[['game_id', 'home_team', 'away_team']].copy()
data1.groupby(['game_id'])['home_team'].count()
results = data1.groupby('game_id').apply(lambda x: x.sample(n=1, random_state=random.seed())).reset_index(drop=True)

```

```python
teams = {team:0 for team in data1.home_team.values}
teams
```

```python
results
```

```python
for col in ['home_team', 'away_team']:
    for value in results[col]:
        if value in teams:
            teams[value] += 1
```

```python
#sort the number of games played by a team in descending order
lst = list(teams.items())
lst = sorted(lst, key=lambda x:x[1], reverse=True)
lst
```
